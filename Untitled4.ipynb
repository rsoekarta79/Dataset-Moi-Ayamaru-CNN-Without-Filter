# **No Filter**

```python
!pip install tensorflow

import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, regularizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications import EfficientNetB1
from tensorflow.keras.applications.efficientnet import preprocess_input
from sklearn.metrics import confusion_matrix
import pathlib
```

```python
# Set path dataset
dataset_url = '/content/drive/MyDrive/datauji'
data_dir = pathlib.Path(dataset_url)

# Hitung jumlah gambar
image_count = len(list(data_dir.glob('*/*.jpg')))
print(image_count)

# Pengaturan batch size dan ukuran gambar
batch_size = 64
img_height = 224
img_width = 224

# Membuat dataset pelatihan dan validasi
train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size)

class_names = train_ds.class_names
print(class_names)

val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size)

# Normalisasi dataset menggunakan preprocess_input
normalized_train_ds = train_ds.map(lambda x, y: (preprocess_input(x), y))
normalized_val_ds = val_ds.map(lambda x, y: (preprocess_input(x), y))

# Data augmentation
data_augmentation = Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
])

# Model dengan EfficientNetB1
base_model = EfficientNetB1(input_shape=(img_height, img_width, 3), include_top=False, weights='imagenet')
base_model.trainable = False

model = Sequential([
    data_augmentation,
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.5),
    layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    layers.Dropout(0.5),
    layers.Dense(len(class_names))
])

# Memanggil model pada satu batch data untuk membangunnya
model.build((None, img_height, img_width, 3))
model.summary()

# Compile model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])


# Train model
epochs = 50
history = model.fit(
    normalized_train_ds,
    validation_data=normalized_val_ds,
    epochs=epochs,
)

# Evaluasi model
def evaluate_model(model, val_ds):
    loss, accuracy = model.evaluate(val_ds)
    print(f"Loss: {loss}, Accuracy: {accuracy}")

print("Evaluasi EfficientNetB1")
evaluate_model(model, normalized_val_ds)

# Visualisasi hasil training
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(range(len(acc)), acc, label='Training Accuracy')
plt.plot(range(len(val_acc)), val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(range(len(loss)), loss, label='Training Loss')
plt.plot(range(len(val_loss)), val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
#Prediksi model pada data validasi
from sklearn.metrics import classification_report
predictions = np.argmax(model.predict(val_ds), axis=-1)

true_labels = []
for _, labels in val_ds:
  true_labels.extend(labels.numpy())
true_labels = np.array(true_labels)

print(classification_report(true_labels, predictions))
```

```python
conf_matrix = confusion_matrix(true_labels, predictions)
unique_labels = np.unique(np.concatenate((true_labels, predictions)))
num_classes = len(unique_labels)

plt.figure(figsize=(8, 6))
plt.imshow(conf_matrix, cmap='Blues')
plt.title('Confusion Matrix')
plt.colorbar()
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.xticks(np.arange(num_classes))
plt.yticks(np.arange(num_classes))
for i in range(num_classes):
    for j in range(num_classes):
        plt.text(j, i, conf_matrix[i, j], ha='center', va='center', color='white')
plt.show()

accuracy = np.sum(np.diag(conf_matrix)) / np.sum(conf_matrix)
print(f'Accuracy: {accuracy}')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf
import pathlib

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from sklearn.metrics import confusion_matrix
from tensorflow.keras.applications import EfficientNetB1
from tensorflow.keras.applications.efficientnet import preprocess_input
```

```python
dataset_url = '/content/drive/MyDrive/datauji'
data_dir = pathlib.Path(dataset_url)
```

```python
image_count = len(list(data_dir.glob('*/*.jpg')))
print(image_count)
```

```python
batch_size = 64
img_height = 224
img_width = 224
```

```python
#Split dataset
train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size)
```

```python
class_names = train_ds.class_names
print(class_names)
```

```python
val_ds = tf.keras.utils.image_dataset_from_directory (
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 15))
for images, labels in train_ds.take(1):
    for i in range(3):
        ax = plt.subplot(5, 5, i + 1)
        # Menampilkan gambar dengan interpolasi bicubic
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
plt.show()
```

```python
for image_batch, labels_batch in train_ds:
  print(image_batch.shape)
  print(labels_batch.shape)
  break
```

```python
# Normalisasi dataset menggunakan preprocess_input
normalized_ds = train_ds.map(lambda x, y: (preprocess_input(x), y))
val_ds = val_ds.map(lambda x, y: (preprocess_input(x), y))
# Membangun model menggunakan EfficientNetB0
num_classes = len(class_names)
efficientnet_base = EfficientNetB1(
    input_shape=(img_height, img_width, 3),
    include_top=False,
    weights='imagenet'
)
# Freeze the base model
efficientnet_base.trainable = False
```

```python
# Build the model
model = Sequential([
    efficientnet_base,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes)
])
```

```python
# Kompilasi model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Pelatihan model
epochs=200
history = model.fit(
    normalized_ds,
    validation_data=val_ds,
    epochs=epochs
)
```

```python
# Visualize training result
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')  # Perubahan disini
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

plt.show()
```

```python
#Prediksi model pada data validasi
from sklearn.metrics import classification_report
predictions = np.argmax(model.predict(val_ds), axis=-1)

true_labels = []
for _, labels in val_ds:
  true_labels.extend(labels.numpy())
true_labels = np.array(true_labels)

print(classification_report(true_labels, predictions))
```

```python
# Menampilkan nilai akurasi menggunakan confusion matrix
conf_matrix = confusion_matrix(true_labels, predictions)

plt.figure(figsize=(8, 6))
plt.imshow(conf_matrix, cmap='Blues')
plt.title('Confusion Matrix')
plt.colorbar()
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.xticks(np.arange(num_classes))
plt.yticks(np.arange(num_classes))
for i in range(num_classes):
    for j in range(num_classes):
        plt.text(j, i, conf_matrix[i, j], ha='center', va='center', color='white')
plt.show()

accuracy = np.sum(np.diag(conf_matrix)) / np.sum(conf_matrix)
print(f'Accuracy: {accuracy}')
```
